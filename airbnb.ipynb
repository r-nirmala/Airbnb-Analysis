```python
import pandas as pd
pd.set_option("display.max_columns",None)
import warnings
warnings.filterwarnings("ignore")
import os
import json
import streamlit as st
from streamlit_option_menu import option_menu
import plotly.express as px
```

```python
#read_airbnb csv file, convert to df
df = pd.read_csv(r"E:\Guvi Data Science - MDE86\Guvi Projects\sample_airbnb.listingsAndReviews.csv")
df.head(3)
```

```python
#view all columns
df.columns
```

Primary datas

```python
#slice specific cols -- primary data
needed_cols = ['_id','listing_url','name','property_type','room_type','bed_type',
                        'minimum_nights','maximum_nights','cancellation_policy','accommodates',
                        'bedrooms','beds','number_of_reviews','bathrooms','price',
                        'cleaning_fee','extra_people','guests_included','images.picture_url',
                        'review_scores.review_scores_rating']
df_pri = df.loc[:,needed_cols]
df_pri.head(2)
```

```python
#check null values
df_pri.isnull().sum()
```

```python
#fill null values
df_pri["beds"].fillna(0,inplace= True)
df_pri["bedrooms"].fillna(0,inplace= True)
df_pri["bathrooms"].fillna(0,inplace= True)
df_pri["cleaning_fee"].fillna(0,inplace= True)
df_pri['name'].fillna(0,inplace=True)
df_pri['review_scores.review_scores_rating'].fillna(0,inplace=True)
```

```python
#rename columns
df_pri.rename(columns ={'review_scores.review_scores_rating':'review_scores'},inplace=True)
df_pri.rename(columns = {'images.picture_url':'images'},inplace=True)
```

```python
df_pri.columns
```

```python
#check null values
df_pri.isnull().sum()
```

```python
df_pri.head(4)
```

```python
df_pri.info()
```

```python
#data type conversion
df_pri['bedrooms'] = df_pri['bedrooms'].astype(int)
df_pri['beds'] = df_pri['beds'].astype(int)
df_pri['bathrooms'] = df_pri['bathrooms'].astype(int)
df_pri['price'] = df_pri['price'].astype(int)
df_pri['cleaning_fee'] = df_pri['cleaning_fee'].astype(int)
df_pri['extra_people'] = df_pri['extra_people'].astype(int)
df_pri['review_scores'] = df_pri['review_scores'].astype(int)
```

```python
df_pri.info()
```

Host data

```python
#view specific cols in df
col = df.columns.tolist()
col[135:150]
```

```python
#slicing needed cols for host
needed_cols = ['_id','host.host_id','host.host_url','host.host_name','host.host_location',
        'host.host_response_time', 'host.host_response_rate','host.host_thumbnail_url',
        'host.host_picture_url','host.host_neighbourhood','host.host_is_superhost',
        'host.host_has_profile_pic','host.host_identity_verified','host.host_listings_count',
        'host.host_total_listings_count']
df_host = df.loc[:,needed_cols]
df_host.head(2)
```

```python
#drop null values in location col
df_host.dropna(subset = ['host.host_location'], inplace = True)
df_host.isnull().sum()
```

```python
df_host.info()
```

```python
#map false and true values to 0 and 1
df_host['host.host_is_superhost']  = df_host['host.host_is_superhost'].map({False: 'No',True : 'Yes'})
df_host['host.host_has_profile_pic']  = df_host['host.host_has_profile_pic'].map({False: 'No',True : 'Yes'})
df_host['host.host_identity_verified']  = df_host['host.host_identity_verified'].map({False: 'No',True : 'Yes'})
```

```python
#fill null values
df_host['host.host_response_rate'].fillna(0,inplace=True)
df_host['host.host_response_time'].fillna('Not Specified',inplace=True)
df_host['host.host_neighbourhood'].fillna('Not Specified',inplace=True)
```

```python
df_host.isnull().sum()
```

```python
df_host.columns
```

```python
#rename column names for host data
df_host.rename(columns={'host.host_id':'host_id','host.host_url':'host_url',
                        'host.host_name':'host_name','host.host_location':'host_location',
                     'host.host_response_time':'host_response_time','host.host_response_rate':'host_response_rate',
                    'host.host_thumbnail_url':'host_thumbnail_url', 'host.host_picture_url':'host_picture_url',
                     'host.host_neighbourhood':'host_neighbourhood','host.host_is_superhost':'host_is_superhost',
                 'host.host_has_profile_pic':'host_has_profile_pic', 'host.host_identity_verified':'host_identity_verified',
                'host.host_listings_count':'host_listings_count','host.host_total_listings_count':'host_total_listings_count'},inplace=True)
```

Amenities

```python
col = df.columns.tolist()
col[25:101]
```

```python
#needed columns
df_amen = df.loc[:,[col[0]]+col[25:101]]
df_amen.head(3)
```

```python
#fill null values of all columns
df_amen.fillna('Not Specified',inplace=True)
```

```python
#merge amenities column and drop single columns
df_amen['amenities'] = df_amen.iloc[:, 1:].apply(lambda row: ','.join(row), axis=1)
df_amen.drop(columns=df_amen.columns[1:-1], inplace=True)
```

```python
#sort the elements
df_amen['amenities_sorted'] = df_amen['amenities'].apply(lambda x: ','.join(sorted(x.split(','))))
df_amen.drop(columns = ['amenities'],inplace = True)
```

```python
#rename column
df_amen.rename(columns={'amenities_sorted':'amenities'},inplace=True)
```

```python
df_amen.head(4)
```

```python
df_amen.isnull().sum()
```

```python
df_amen.info()
```

```python
#merge all dataframes
#df_pri
#df_host
#df_addr
#df_avail
#df_amen

df = pd.merge(df_pri,df_host, on ='_id')
df = pd.merge(df,df_addr,on='_id')
df= pd.merge(df, df_avail, on='_id')
df = pd.merge(df, df_amen, on = '_id')
df.head(2)
```

```python
#check for null in final df
df.isnull().sum()
```
